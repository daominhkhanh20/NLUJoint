```python
data = pd.read_csv('../assets/data/bkai/train/train.csv')

data_relabel = pd.read_csv('../assets/data/bkai/train/relabel_train.csv')
```

```python
data['text_copy'] = data['text'].astype(str).progress_apply(lambda x: preprocess_sent(x))
```

```python
data['tokenizer_sent'] = data['text_copy'].astype(str).apply(lambda x: ViTokenizer.tokenize(x))
```

```python
corpus_sent_by_label = dict()
for idx, row in data.iterrows():
    if row['intent'] not in corpus_sent_by_label:
        corpus_sent_by_label[row['intent']] = {
            'text': f"{row['tokenizer_sent']} ",
            'count': 1
        }
    else:
        corpus_sent_by_label[row['intent']]['text'] += f"{row['tokenizer_sent']} "
        corpus_sent_by_label[row['intent']]['count'] += 1
        
```

```python
top_word_common_by_label = dict()
for key, value in corpus_sent_by_label.items():
    counter_word = Counter(value['text'].split(" "))
    
    top_words = counter_word.most_common(20)
    top_word_common_by_label[key] = {}
    for word, word_counter in top_words:
        if word != 'phần_trăm':
            top_word_common_by_label[key][word] = word_counter / value['count']
        else:
            top_word_common_by_label[key][word] = 2 * word_counter / value['count']
    # top_word_common_by_label[key] = {pair[0]: pair[1] if pair[0] != 'phần trăm' else pair[0]: pair[1]  for pair in top_words}
```

```python
top_word_common_by_label
```

```python
def assign_label(sent, debug=False):
    sent = ViTokenizer.tokenize(sent).split(" ")
    # print(sent)
    
    label = None
    score = 0
    for key, value in top_word_common_by_label.items():
        temp = {}
        temp_score = 0
        # temp_appear = 0
        for word in sent:
            if word in value:
                temp[word] = value[word]
                temp_score += value[word]
                # temp_appear += 1
        
        if temp_score > score:
            score = temp_score
            label = key
        # if temp_appear!= 0 and temp_score / temp_appear > score:
        #     score = temp_score / temp_appear
        #     label = key
        if debug:
            print(key)
            print(temp)
            print(temp_score)
        # if temp_appear!=0:
        #     print(temp_score / temp_appear)
        # print()
    
    return label, score
        
```

```python
assign_label('hãy tăng giúp ta cái nóng lạnh 4 ở phòng trẻ em 5', debug=True)
```

```python
new_labels = []
is_diff = []
# cnt_old_labels, cnt_new_labels, cnt_relabels = [], [], []
for idx, row in tqdm(data.iterrows(), total=len(data)):
    sent = row['text']
    label, score = assign_label(sent)
    new_labels.append(label)            
    if label != row['intent']:
        is_diff.append(True)
    else:
        is_diff.append(0)
```
